# Melt()

```python
import pandas as pd
var=pd.DataFrame({"Days":[1,2,3,4,5,6],"Maths":[56,77,67,87,67,88],"English":[89,99,98,97,96,66]})
var
```

```python
pd.melt(var)
```

```python
pd.melt(var,id_vars=["English"],var_name="Python",value_name="Marks")
```

# pivot()

```python
var=pd.DataFrame({"Days":[1,2,3,4,5,6],
                  "st_name":["a","b","c","d","e","f"],
                  "Maths":[56,77,67,87,67,88],
                  "English":[89,99,98,97,96,66]
                 })
var
```

```python
var.pivot(index="Days",columns="st_name")
```

```python
var.pivot(index="Days",columns="st_name",values="English")
```

```python
var=pd.DataFrame({"Days":[1,1,1,1,2,2],
                  "st_name":["a","b","c","d","e","f"],
                  "Maths":[56,77,67,87,67,88],
                  "English":[89,99,98,97,96,66]
                 })
var
```

```python
var.pivot_table=(index=="st_name",columns=="Days",aggfunc=="sum",margins==True)
```
