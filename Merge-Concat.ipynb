# Merge


```python
import pandas as pd

csv1=pd.DataFrame({"A":[1,2,3,4,5],"B":[15,12,23,43,55]})
csv2=pd.DataFrame({"A":[1,2,3,4,5],"C":[1,1,2,4,5]})
pd.merge(csv1,csv2,on="A")
```

```python
csv1=pd.DataFrame({"A":[1,2,3,4],"B":[15,12,23,43]})
csv2=pd.DataFrame({"A":[1,2,3,5],"C":[1,1,2,4]})
pd.merge(csv1,csv2,how="left")
```

```python
csv1=pd.DataFrame({"A":[1,2,3,4],"B":[15,12,23,43]})
csv2=pd.DataFrame({"A":[1,2,3,5],"B":[1,1,2,4]})
pd.merge(csv1,csv2,left_index=True,right_index=True,suffixes=("_Name","_Id"))
```

# concate

```python
sr1=pd.Series([1,2,3,4,5])
sr2=pd.Series([11,21,31,41,51])
pd.concat([sr1,sr2])
```

```python
sr1=pd.DataFrame({"A":[1,2,3,4],"B":[55,66,77,88]})
sr2=pd.DataFrame({"A":[1,2,3,4],"B":[11,22,33,44]})
pd.concat([sr1,sr2])
```

```python
sr1=pd.DataFrame({"A":[1,2,3,4],"B":[55,66,77,88]})
sr2=pd.DataFrame({"A":[1,2,3,4],"B":[11,22,33,44]})
pd.concat([sr1,sr2],axis=0)
```

```python
sr1=pd.DataFrame({"A":[1,2,3,4],"B":[55,66,77,88]})
sr2=pd.DataFrame({"A":[1,2,3,4],"B":[11,22,33,44]})
pd.concat([sr1,sr2],axis=1)
```

```python
sr1=pd.DataFrame({"A":[1,2,3,4],"B":[55,66,77,88]})
sr2=pd.DataFrame({"A":[1,2],"B":[11,22]})
pd.concat([sr1,sr2],axis=1,join="outer")
```

```python
sr1=pd.DataFrame({"A":[1,2,3,4],"B":[55,66,77,88]})
sr2=pd.DataFrame({"A":[1,2],"B":[11,22]})
pd.concat([sr1,sr2],axis=1,join="inner")
```

```python
sr1=pd.DataFrame({"A":[1,2,3,4],"B":[55,66,77,88]})
sr2=pd.DataFrame({"A":[1,2],"B":[11,22]})
pd.concat([sr1,sr2],axis=1,keys=["sr1","sr2"])
```

```python
sr1=pd.DataFrame({"A":[1,2,3,4],"B":[55,66,77,88]})
sr2=pd.DataFrame({"A":[1,2],"B":[11,22]})
pd.concat([sr1,sr2],axis=0,keys=["sr1","sr2"])
```
